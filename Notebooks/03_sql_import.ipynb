```python
import pandas as pd
from sqlalchemy import create_engine

# Load CSV
df = pd.read_csv(
    r"C:\Users\khyat\OneDrive\Desktop\Amazon Supply Chain Intelligence & Demand Forecasting\Data\clean_supply_chain.csv"
)

# MySQL Connection
engine = create_engine(
    "mysql+pymysql://root:root@localhost:3306/amazon_supply_chain"
)

# Export to MySQL
df.to_sql(
    name="supply_chain",
    con=engine,
    if_exists="replace",
    index=False
)

print("✅ Data imported successfully!")
print("Total Rows:", len(df))
```
